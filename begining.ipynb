```python
#all temp 10, 25,40,50,25,20 for all set of sample
def alltemp( data10, data25, data40, data50, data25ah, data20ah):
    dat=[data25, data10, data40, data50, data25ah, data20ah]
    df=[]
    for i in range(len(dat)):
        df.append(np.genfromtxt(dat[i], delimiter='\t',skip_header=12))
    df=np.array(df)
    q = df[:,:,0]  #size (6, 234, 3) so [:,:,0] gives q
    I = df[:,:,1]
    color=['red','blue','green','white','cyan','yellow']
    lev=['10K','25K','40K','50K','25K_ah','20K_ah']
    plt.style.use('dark_background')
    fig, ax= plt.subplots(figsize=(20,8))
    ax.set(ylabel='Intensity', xlabel='Q(A-1)')
    for i in range(len(dat)):
        fig=ax.plot(q[i,:],I[i,:], color=color[i],label=lev[i])
    ax.legend()
    ax.set_xlim([0.035, 1])
    ax.set_ylim([0, 2.5])
    return fig
    
```

```python
def alltemp_s( data10, data25, data40, data50, data25ah, data20ah, 
            data_ep10, data_ep25, data_ep40, data_ep50, data_ep25ah, data_ep20ah):
    dat=[data25, data10, data40, data50, data25ah, data20ah]
    df=[]
    for i in range(len(dat)):
        df.append(np.genfromtxt(dat[i], delimiter='\t',skip_header=12))
    df=np.array(df)
    q = df[:,:,0]  #size (6, 234, 3) so [:,:,0] gives q
    
    df_ec=[]
    dat_ec=[data_ep10, data_ep25, data_ep40, data_ep50, data_ep25ah, data_ep20ah]
    for i in range(len(dat_ec)):
        df_ec.append(np.genfromtxt(dat_ec[i], delimiter='\t',skip_header=12))
    df_ec=np.array(df_ec)
    
    I=[]
    for i in range(len(dat_ec)):
        I.append(df[i,:,1]-df_ec[i,:,1])   # substracting each temp empty capillary
    I=np.array(I)
    
    color=['red','blue','green','white','cyan','yellow']
    lev=['10K','25K','40K','50K','25K_ah','20K_ah']
    plt.style.use('dark_background')
    fig, ax= plt.subplots(figsize=(20,8))
    ax.set(ylabel='Intensity', xlabel='Q(A-1)')
    for i in range(len(dat)):
        fig=ax.plot(q[i,:],I[i,:], color=color[i],label=lev[i])
    ax.legend()
    ax.set_xlim([0.035, 1])
    ax.set_ylim([0, 2.5])
    return fig
```

```python
#substracting empty cappilary(dataec)
def sametemp_s1( data1, data2, data3, dataec):
    dat=[data1, data2, data3]
    df=[]
    for i in range(len(dat)):
        df.append(np.genfromtxt(dat[i], delimiter='\t',skip_header=12))
    df=np.array(df)
    df_ec=np.genfromtxt(dataec, delimiter='\t',skip_header=12) #for empty capilliry
    
    q = df[:,:,0]  #size (6, 234, 3) so [:,:,0] gives q
    I=df[:,:,1]-df_ec[:,1]
    
    color=['red','blue','green']
    lev=['DP1','DP2','DP3']
    plt.style.use('dark_background')
    fig, ax= plt.subplots(figsize=(20,8))
    ax.set(ylabel='Intensity', xlabel='Q(A-1)')
    for i in range(len(dat)):
        fig=ax.plot(q[i,:],I[i,:], color=color[i],label=lev[i])
    ax.legend()
    ax.set_xlim([0.035, 1])
    ax.set_ylim([0, 2.5])
    return fig
```
