```python
#creating folder
vfolder=Path.cwd()/'vertical'   #folder for vertical cut
hfolder=Path.cwd()/'horizantal'  #folder for horizontal cut
sec_folder=Path.cwd()/'Sunil_Sinha_Nov2019' # folder for sector cut(0,30,60,90,120,150 degree)
```

```python
#all data for different sector
igor_2dfig=Path.cwd()/'2d_figures'  # 2d figures folder
vert_comb_low='V*comb*low*.ABS'  #vertical cut for all temp at 700 Oe
vert_comb_1='V*comb*1*.ABS'  #vertical cut for all temp at 1 kOe
vert_comb_2='V*comb*2*.ABS'  #vertical cut for all temp at 2 kOe

horz_comb_low='H*comb*low*.ABS' #horizontal cut for all temp at 700 Oe
horz_comb_1='H*comb*1*.ABS'   #horizontal cut for all temp at 1 kOe
horz_comb_2='H*comb*2*.ABS' #horizontal cut for all temp at 2 kOe

vert_comb=['V18K*comb*.ABS','V20K*comb*.ABS','V22K*comb*.ABS','V24K*comb*.ABS','V25K*comb*.ABS',
           'V26K*comb*.ABS','V30K*comb*.ABS','V75K*comb*.ABS', 'V200K*comb*.ABS']
horz_comb=['H18K*comb*.ABS','H20K*comb*.ABS','H22K*comb*.ABS','H24K*comb*.ABS','H25K*comb*.ABS',
           'H26K*comb*.ABS','H30K*comb*.ABS','H75K*comb*.ABS','H200K*comb*.ABS'] 
sec_15m=['15m_18K*deg*.ASC','15m_20K*deg*.ASC','15m_24K*deg*.ASC','15m_26K*deg*.ASC','15m_30K*deg*.ASC']
```

```python
#importing SANS class, which take folowing value ;(data, xrange=[0.001,0.07], yrange=[0.0001,3000], folder)

vert_low=SANS(vert_comb_low,folder=vfolder)
vert_1=SANS(vert_comb_1, folder=vfolder)
vert_2=SANS(vert_comb_2, folder=vfolder)

horz_low=SANS(horz_comb_low, folder=hfolder)
horz_1=SANS(horz_comb_1, folder=hfolder)
horz_2=SANS(horz_comb_2, folder=hfolder)

vert_all=SANS(vert_comb,folder=vfolder )
horz_all=SANS(horz_comb,folder=hfolder )
sec_plot=SANS(sec_15m, xrange=[0.001, 0.01],yrange=[0.1, 3000],folder=sec_folder)
```
